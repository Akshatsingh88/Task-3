### Our next step is to find the relationship

```python
cor = df.corr()
```

```python
sns.heatmap(cor , xticklabels = cor.columns , yticklabels = cor.columns , annot = True)
```

### Here we can see that postal code has almost no relation with other parameters so we should drop it

```python
df1 = df.drop(columns=["Postal Code" ,"Country"], axis = 1)
df1
```

```python
cor = df1.corr()
sns.heatmap(cor , xticklabels= cor.columns , yticklabels = cor.columns , annot = True)
```

### 1.HERE WE OBSERVW THAT 'PROFIT' AND 'QUANTITY' ARE LEAST CORRELATED AND 
### ALSO 'QUANTITY' AND 'DISCOUNT' ARE ALMOST UNCORELATED

### 2.MAXIMUM CORRELAYION BETWEEN 'SALES' AND 'PROFIT'
